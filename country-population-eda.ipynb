<h3> Additionally, data on income groups is to be separated. <br>
    By international standards, countries can be divided into:
    <ul>
        <li> low income </li>
        <li> lower middle income </li>
        <li> upper middle income </li>
        <li> high income </li>
    </ul>
    Our data includes a category called 'middle income', which is derived from lower and upper middle income. This field has been ignored </h3>

```python
income_groups = ["Low income", "Lower middle income", 
                 "Upper middle income", "High income"]
```

```python
df['Country Name'] = df['Country Name'].map(countries_renamed).fillna(df['Country Name'])
```

```python
country_names = country_names.tolist()
country_names.extend(countries_renamed.values())
len(country_names)
```

```python
df2 = df.drop('Country Name', axis=1).copy()
df2 = df2.transpose()
df2.columns = df['Country Name']
df2.head()
```

```python
df_countries = df2[country_names].copy()
df_countries = df_countries.reindex(sorted(df_countries.columns), axis=1)
df_countries.head()
```

# Analysis

```python
fig1 = px.line(df_countries, x=df_countries.index, y=df_countries.columns,
               labels={
                   "index": "Year",
                   "value": "Population",
               },
              )
fig1.update_layout(title_text='Population Trend Across Countries', title_x=0.5)
fig1.show()
```

```python
def update_plot_country(year):
    modified_data = df_countries.loc[str(year)].copy()
    fig2 = px.pie(modified_data, values=modified_data, names=modified_data.index)
    fig2.update_traces(textposition='inside')
    fig2.update_layout(title_text='World Population Pie', title_x=0.5,
                       uniformtext_minsize=12, uniformtext_mode='hide')
    fig2.show()

interact(update_plot_country, year = widgets.IntSlider(
    value = 1960,
    min=1960,
    max=2022,
    step=1,
    description="Year:"
))
```

```python
df_income = df2[income_groups].copy()
```

```python
fig3 = px.area(df_income, x=df_income.index, y=df_income.columns,
               labels={
                   "Country Name": "Income Group",
                   "index": "Year",
                   "value": "Population",
               },
              )
fig3.update_layout(title_text='Population Trend Across Income Groups', title_x=0.5)
fig3.show()
```

```python
def update_plot_income(year):
    modified_data = df_income.loc[str(year)].copy()
    fig4 = px.pie(modified_data, values=modified_data, names=modified_data.index,
                  labels={
                      "Country Name": "Income Group",
                      "index": "Year",
                      "value": "Population"
                  }
                 )
    fig4.update_traces(textposition='inside')
    fig4.update_layout(title_text='World Population Pie', title_x=0.5,
                       uniformtext_minsize=12, uniformtext_mode='hide')
    fig4.show()
    

interact(update_plot_income, year = widgets.IntSlider(
    value = 1960,
    min=1960,
    max=2022,
    step=1,
    description="Year:"
))
```

# Regression

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
df_pred = pd.DataFrame(columns=df_countries.columns)
pred_years = np.arange(1960, 2031).reshape(-1, 1)

x = np.array(list(map(int, df2.index))).reshape(-1, 1)

poly = PolynomialFeatures(3, include_bias=False)

for country in df_pred.columns:
    y = np.array(df2[country]).reshape(-1, 1)

    poly_features = poly.fit_transform(x)
    
    model = LinearRegression()
    model.fit(poly_features, y)
    
    pred_features = poly.fit_transform(pred_years)
    y_pred = model.predict(pred_features)
    df_pred[country] = y_pred.squeeze()

df_pred.set_index(pred_years.squeeze(),  inplace=True)
```

```python
df_pred.head()
```

```python
fig1 = px.line(df_pred, x=df_pred.index, y=df_pred.columns,
               labels={
                   "index": "Year",
                   "value": "Predicted Population",
               },
              )
fig1.update_layout(title_text='Approx and Predicted Population Across Countries', title_x=0.5)
fig1.show()
```
